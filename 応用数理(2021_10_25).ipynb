###**力学系 (dynamical system)**###
$$\frac{dx}{dt}=f(x,y)\tag{1}$$
$$\frac{dy}{dt}=g(x,y)\tag{2}$$
ここで、$f(x,y), g(x,y)$は、$x,y$の連続関数である。
(*式番号はGitHub上では表示されませんが、Google Colabでは表示されます*)

###**相平面上の解軌道**###
初期条件$x(0)=x_0, y(0)=y_0$に対して、式(1)(2)は、$t\geq0$で一意に解$x=\phi(t), y=\psi(t)$を持つ。また、$t$を媒介変数として、相平面とよばれる$x-y$平面上に、解の曲線を表すことができ、これを軌道という。
解曲線の傾きは以下で与えられる。
$$\frac{dy}{dx}=\frac{dy/dt}{dx/dt}=\frac{g(x,y)}{f(x,y)}$$
$f(x_c,y_c)=g(x_c,y_c)=0$となる点$(x_c,y_c)$は平衡点または臨界点と呼ばれる。この点では、式(1)(2)の右辺は0となるため、軌道は$x=x_c,y=y_c$にとどまる。
軌道の振る舞いには以下の性質がある。
1.   相平面内の任意の点を通る軌道は存在するなら１つである
2.   平衡点でない点から出発する軌道は、有限時間で平衡点に到達できない。ある軌道が平衡点$(x_c,y_c)$に近づくなら、$t\rightarrow\infty$となる
3.  軌道は閉曲線でなければ、自分自身と交わらない。閉曲線は周期解に対応する。




###**平衡点近傍の振る舞い**###
任意の平衡点は、相平面上の座標変換（平行移動）で原点に移せるので、平衡点を$(x_c,y_c)=(0,0)$とする。関数$f(x,y)$と$g(x,y)$に対して、平衡点周りで級数展開する。
点$(0,0)$の$(n+m)$階の微分係数$$\frac{\partial^{n+m} f}{\partial^n x\partial^m y}(0,0)=\frac{\partial^{n+m} f}{\partial^n x\partial^m y}|_{x=0,y=0}$$と表すと、
$$\frac{dx}{dt}=f(x,y)=f(0,0)+\frac{\partial f}{\partial x}(0,0)x+\frac{\partial f}{\partial y}(0,0)y+\cdots$$
$$\frac{dy}{dt}=g(x,y)=g(0,0)+\frac{\partial g}{\partial x}(0,0)x+\frac{\partial g}{\partial y}(0,0)y+\cdots$$
高次項を無視することで、以下の連立微分方程式を得る。
$$\frac{dx}{dt}=ax+by$$
$$\frac{dy}{dt}=cx+dy$$
ここで、$a:=\frac{\partial f}{\partial x}(0,0)$,
$b:=\frac{\partial f}{\partial y}(0,0)$,
$c:=\frac{\partial g}{\partial x}(0,0)$,
$d:=\frac{\partial g}{\partial y}(0,0)$
一般解は
$$x=A_1\exp(r_1t)+A_2\exp(r_2t)$$
$$x=B_1\exp(r_1t)+B_2\exp(r_2t)$$
ただし、$r_1$, $r_2$は、$$r^2-(a+d)r+ad-bc=0$$の解。
$r_1$、$r_2$の値により、以下の平衡点の性質が決まる。
1.   **安定**: $t\rightarrow\infty$のとき$x$, $y$が有界
2.   **漸近的安定**: $t\rightarrow\infty$のとき$(x,y)\rightarrow (0,0)$
3.   **不安定**: その他



##**生物間の相互作用**##
###捕食者と被食者###
生物種は、様々な方法で、互いに影響を及ぼし合っている。ここでは捕食者(predator)と被食者(prey)との関係についての有名な数理モデルであるLotka-Volterra方程式を学ぼう。
捕食者の個体数$y$、被食者の個体数$x$とする。式(1)(2)において、
$$f(x,y)=ax-bxy$$
$$g(x,y)=-cx+dxy$$
のとき、Lotka-Volterra方程式となる。

$$\frac{dx}{dt}=ax-bxy\tag{3}$$
$$\frac{dy}{dt}=-cx+dxy\tag{4}$$

捕食者がいないときには($y=0$)、被食者の個体数は$dx/dt=ax$で際限なく増える。また、被食者がいなければ($x=0$)、捕食者の個体数は$dy/dt=-cy$で減る。

捕食者と被食者の相互作用は、その遭遇率に依存すると考え、$xy$に定数比例した項を導入する。捕食者は被食者を得ることで増える(dxy>0)ことができ、被食者は減る(-bxy<0)。

式(3)(4)の右辺0の条件から、平衡点$(0,0)$と$(c/d,a/b)$があることが分かる。これらは、それぞれ、鞍点と渦心点である（演習）。

周期解のとき、周期$T$での個体数の平均値を求める。
$$\bar x=\frac{1}{T}\int_0^T x(t)dt$$
$$\bar y=\frac{1}{T}\int_0^T y(t)dt$$
式(3)より
$$\frac{1}{x}\frac{dx}{dt}=a-by$$
の両辺を積分すると
$$\log(x(T)/x(0))=aT-b\int_0^T y(t)dt$$
いま、周期$T$の周期解を考えているので、$x(T)=x(0)$である。
したがって、
$$0=aT-b\int_0^T y(t)dt$$となり
$$\bar y = \frac{a}{b}$$
同様に$$\bar x = \frac{c}{d}$$となる。
このように、平均値は周期解の渦心点と一致する。

###**演習**###
平衡点$(0,0)$と$(c/d,a/b)$は、それぞれ、鞍点と渦心点になることを確かめよ。



###***プログラム***###

```python
%matplotlib inline
import numpy as np
from scipy.integrate import odeint
from scipy import integrate
import math
import matplotlib.pyplot as plt


def func(v, t, a, b, c, d):
  """
  Prey: x=v[0]
  Predator: y=v[1]
  dx/dt = ax-bxy (a>0, b>0)
  dy/dt = -cy+dxy (c>0, d>0)

  Prey-Predator interaction (encounter rate) as xy
  Prey increases as ax and decreases as bxy
  Predator decreases as cy and increases as dxy
  """
  return [a*v[0]-b*v[0]*v[1], -c*v[1]+d*v[0]*v[1]]

a=1
b=1
c=1
d=1
T=100
v0 = [0.1, 0.1]
t = np.arange(0, T, 0.01)
v = odeint(func, v0, t, args=(a,b,c,d))

x= integrate.simps(v[:,0],t)/T
y= integrate.simps(v[:,1],t)/T
print("(numerical) x_mean =",round(x,3)," y_mean =",round(y,3))

plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
plt.plot(v[:, 0], v[:, 1],label="trajectory")
plt.legend()
plt.subplot(1,2,2)
plt.plot(t,v[:, 0],label="prey")
plt.plot(t,v[:, 1],label="predator")
plt.legend()
```

###***人間の活動の影響***###
Lotka-Volterra方程式(3)(4)に、人間の活動の影響を取り込もう。例えば、漁業であれば、捕食者はサメ、被食者は食用魚であり、魚網で一網打尽にこれらを獲るとする。漁獲の影響を$E$であらわすと
$$\frac{dx}{dt}=ax-bxy-Ex$$
$$\frac{dy}{dt}=-cx+dxy-Ey$$
である。これらは、渦心点を
$$(\frac{c}{d},\frac{a}{b})\rightarrow(\frac{c+E}{d},\frac{a-E}{b})$$
にずらす。これは適正な漁獲$E<a$が実際には食用魚（被食者）の個体数を増加させ、サメ（被食者）の平均個体数を減らすことを意味する。逆に、漁獲を止めると、平均的にはサメの数は増え、食用魚の数が減少する。

もちろん、これは漁獲に限らない。殺虫剤、農薬など、個体数を減少させる効果のある活動を捕食者―被食者関係の生態系に作用させると、同様の結果が生じると期待できる。（害虫駆除のつもりが、その天敵を減らして、結果的に害虫が増える）

このように、非線形な微分方程式系からは、直感から期待されることとは異なる結果が生じることがある。

##**競争関係（資源の奪い合い）**##
人口増加の数理モデルで、環境収容力を学んだ。この環境収容力（資源）を奪い合うような競争関係が２種の生物間（国家間、企業間でも良い）に存在する場合の生態系を考える。
$$\frac{dN_1}{dt}=r_1N_1(1-\frac{N_1}{K_1}-\frac{\alpha_{12}N_2}{K_1})\tag{5}$$
$$\frac{dN_2}{dt}=r_2N_2(1-\frac{N_2}{K_2}-\frac{\alpha_{21}N_1}{K_2})\tag{6}$$
式(5)において、$N_2=0$とすると、人口増加の数理モデル（第1回講義）と同じ形式であることが分かるだろう。つまり$K$は環境収容力である。

例えば、2種の競争力を比べたときに、もし、$\alpha_{12}/K_1 < \alpha_{21}/K_2$であれば、生物種1のほうが生物種2よりも**強い**ということになる。

$\frac{dN_1}{dt}=\frac{dN_2}{dt}=0$の条件から平衡点を求めると、
$(0,0)$, $(0,K_2)$, $(K_1,0)$,
$(\frac{K_1-\alpha_{12}K_2}{1-\alpha_{12}\alpha_{21}},\frac{K_2-\alpha_{21}K_1}{1-\alpha_{12}\alpha_{21}})$の4点である。ただし4番目の平衡点は、条件$1-\alpha_{12}\alpha_{21}\neq0$が成立する場合に存在。

###**例1**###
$r_1=r_2>0$, $K_1=K_2=K$, $\alpha_{21}>\alpha_{12}>1$を考える。これは、種内競争の条件は同じで、種間競争では生物種1のほうが強い場合である。
この場合、


1.   $(0,0)$
2.   $(0,K)$
3.   $(K,0)$
4.   $(\frac{1-\alpha_{12}}{1-\alpha_{12}\alpha_{21}}K,\frac{1-\alpha_{21}}{1-\alpha_{12}\alpha_{21}}K)$

のうち、平衡点1，4は不安定、平衡点2,3が安定となる。
平衡点4が不安定であることから、この場合は共存することは起こりえず、どちらか一方が死滅する（平衡点2または3）。平衡点2，3のどちらに行くかは、$(N_1,N_2)$の初期条件で決まる。

###**例2**###
$\frac{K_1}{\alpha_{12}}>K_2$かつ$\frac{K_2}{\alpha_{21}}>K_1$のとき、平衡点4は安定となり、種1と2は共存する。
$r_1, r_2>0$,
$K_1=K_2=K$であれば、$\alpha_{12}, \alpha_{21} < 1$が条件である。

（数値計算をせずに、軌道の概略図を描くやり方を板書で説明する）



###**プログラム**###


```python
%matplotlib inline
import numpy as np
from scipy.integrate import odeint
from scipy import integrate
import math
import matplotlib.pyplot as plt


def func(v, t, r_1,r_2, a_12,a_21,K_1,K_2):
  """
  species1: N_1=v[0]
  species2: N_2=v[1]
  dx/dt = r_1N_1(1-(N_1+a_12N_2)/K_1)
  dy/dt = r_2N_2(1-(N_2+a_21N_1)/K_2)
  """
  return [r_1*v[0]*(1-(v[0]+a_12*v[1])/K_1), r_2*v[1]*(1-(v[1]+a_21*v[0])/K_2)]

r_1=1
r_2=1
K_1=1
K_2=1

a_12=2
a_21=3

if abs(1-a_12*a_21) > 1e-5:
  print( "4th point (", (K_1-a_12*K_2)/(1-a_12*a_21),",",(K_2-a_21*K_1)/(1-a_12*a_21),")"  )

T=100
v0 = [0.3,0.3]
t = np.arange(0, T, 0.01)
v = odeint(func, v0, t, args=(r_1,r_2,a_12,a_21,K_1,K_2))


plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
plt.plot(v[:, 0], v[:, 1],label="trajectory")
plt.legend()
plt.subplot(1,2,2)
plt.plot(t,v[:, 0],label="Spc. 1")
plt.plot(t,v[:, 1],label="Spc. 2")
plt.legend()
```
